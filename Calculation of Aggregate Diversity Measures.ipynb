## Winsorising my values to tackle potential outliers

### Determining whether to use 95th or 99th percentile

```python
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt

# Create a copy of the DataFrame if not already done
df_winsorized = df_to_analyse.copy()

# List of variables to Winsorize
variables = ['danceability', 'energy', 'speechiness', 
             'acousticness', 'instrumentalness', 'liveness', 
             'valence', 'track_popularity', 'artist_popularity']

# Applying Winsorization
for var in variables:
    df_winsorized[f'{var}_95'] = winsorize(df_winsorized[var], limits=[0, 0.05])
    df_winsorized[f'{var}_99'] = winsorize(df_winsorized[var], limits=[0, 0.01])

# Setting up the figure to fit on a smaller format (approximately A5)
fig, axes = plt.subplots(nrows=len(variables), ncols=3, figsize=(8.27, 11.69))

# Plotting each variable
for i, var in enumerate(variables):
    # Original data
    axes[i, 0].hist(df_winsorized[var], bins=30, color='blue', alpha=0.7)
    axes[i, 0].set_title(f'Original Histogram of {var}', fontsize=8)
    
    # Winsorized at 95%
    axes[i, 1].hist(df_winsorized[f'{var}_95'], bins=30, color='green', alpha=0.7)
    axes[i, 1].set_title(f'Histogram of {var} Winsorized at 95%', fontsize=8)

    # Winsorized at 99%
    axes[i, 2].hist(df_winsorized[f'{var}_99'], bins=30, color='red', alpha=0.7)
    axes[i, 2].set_title(f'Histogram of {var} Winsorized at 99%', fontsize=8)

    # Adjusting the font size of axis labels for better fit
    for ax in axes[i]:
        ax.tick_params(axis='both', which='major', labelsize=6)

# Adjusting layout to fit on a smaller format
plt.tight_layout(pad=0.5)
plt.show()
```

__Conclusion:__ 
- Winsorization at the 99th percentile is generally the preferred method for managing outliers across various musical and popularity attributes in the dataset.
- This approach effectively reduces the impact of extreme values while preserving the essential distribution and characteristics of the data, crucial for nuanced analysis.

This level of Winsorization helps in managing the skewness while keeping the analysis sensitive to variations in streaming performance across a broader range of tracks.

## Diversity Measure 2: Artist and Track Popularity spread

- I have 2 measures of popularity in my df: track_popularity and artist popularity
- Track popularity is based on the total number of plays the track has had and how recent those plays are
- Artist's popularity is calculated from the popularity of all the artist's tracks
- First let's use the Gini coefficient and calculate it both on the track and the artist popularity levels


```python
import numpy as np

def gini_coefficient(array):
    """Calculate the Gini coefficient of a numpy array."""
    # All values are sorted and normalized.
    array = np.array(array, dtype=np.float64)
    array = array.flatten()  # Flatten the array if not already 1D.
    if np.amin(array) < 0:
        array -= np.amin(array)  # Shift the array to be positive if necessary.
    total = np.sum(array)
    if total == 0:
        return 0  # Return 0 if array sums to 0 (no inequality if there is no presence).
    array = np.sort(array)  # Sort array
    index = np.arange(1, array.shape[0] + 1)  # Create index array
    # Gini coefficient formula
    return (np.sum((2 * index - array.shape[0] - 1) * array)) / (array.shape[0] * np.sum(array))

# Group by country and date and calculate Gini coefficient for track_popularity
df_to_analyse['Gini_coefficient_track_popularity'] = df_to_analyse.groupby(['country', 'date'])['track_popularity'].transform(lambda x: gini_coefficient(x))

# Calculate Gini coefficient for artist popularity
df_to_analyse['Gini_coefficient_artist_popularity'] = df_to_analyse.groupby(['country', 'date'])['artist_popularity'].transform(lambda x: gini_coefficient(x))

display(df_to_analyse.head())
```

### Add annual Gini coefficient calculations for visualisation

```python
# Ensure the 'date' column is in datetime format
df_to_analyse['date'] = pd.to_datetime(df_to_analyse['date'])

# Extract year from the 'date'
df_to_analyse['year'] = df_to_analyse['date'].dt.year

# Define a function to apply the Gini calculation for groups
def calculate_yearly_gini(df, column_name):
    return df.groupby(['country', 'year'])[column_name].apply(gini_coefficient)

# Calculate yearly Gini coefficient for track popularity
gini_coefficient_track_yearly = calculate_yearly_gini(df_to_analyse, 'track_popularity')

# Calculate yearly Gini coefficient for artist popularity
gini_coefficient_artist_yearly = calculate_yearly_gini(df_to_analyse, 'artist_popularity')

# Combine results into a single DataFrame
yearly_gini_results = pd.DataFrame({
    'Gini_coefficient_track_popularity': gini_coefficient_track_yearly,
    'Gini_coefficient_artist_popularity': gini_coefficient_artist_yearly
}).reset_index()

display(yearly_gini_results)
```

```python
import numpy as np
import pandas as pd

# Assuming df_to_analyse is your original DataFrame and it includes a 'genres' column with lists of genres.

# Step 1: Explode the DataFrame so each genre gets its own row
df_exploded = df_to_analyse.explode('genres')

# Step 2: Group by 'country', 'date', and 'genres' to count occurrences
genre_counts = df_exploded.groupby(['country', 'date', 'genres']).size().reset_index(name='count')

# Step 3: Calculate total genres per 'country' and 'date'
total_genres = genre_counts.groupby(['country', 'date'])['count'].sum().reset_index(name='total')

# Step 4: Merge the counts back to calculate proportions
merged = pd.merge(genre_counts, total_genres, on=['country', 'date'])

# Step 5: Calculate proportion p_i
merged['proportion'] = merged['count'] / merged['total']

# Step 6: Calculate Shannon Diversity Index terms
merged['term'] = merged['proportion'] * np.log(merged['proportion'])

# Step 7: Sum the terms grouped by 'country' and 'date' to get Shannon Diversity Index
diversity = merged.groupby(['country', 'date'])['term'].sum().reset_index()
diversity['Shannon_diversity'] = -diversity['term']

# Step 8: Merge the Shannon Diversity Index back to the original DataFrame
df_to_analyse = pd.merge(df_to_analyse, diversity[['country', 'date', 'Shannon_diversity']],
                         on=['country', 'date'], how='left')

# Rename the 'Shannon_diversity' column for clarity
df_to_analyse.rename(columns={'Shannon_diversity': 'Shannon\'s Genre Diversity'}, inplace=True)

# Check the result
display(df_to_analyse.head())
```

### Unweighted Z scores all numeric audio features (rejected)

#### Step 1: Standardise audio features to Z-scores

```python
import pandas as pd
import numpy as np

# Assuming df_to_analyse is your initial DataFrame with all the data
# List of specific audio feature columns
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 
            'loudness', 'speechiness', 'valence', 'tempo', 'duration_ms']

# Standardize features to Z-scores
df_to_analyse[features] = df_to_analyse[features].apply(
    lambda x: (x - x.mean()) / x.std(), axis=0
)
```

### Unweighted Z-scores per country per day

```python
import pandas as pd
import numpy as np

# Assuming df_to_analyse is your DataFrame with the relevant data
# List of specific audio feature columns
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence']

# Ensure 'date' column is in datetime format
df_to_analyse['date'] = pd.to_datetime(df_to_analyse['date'], errors='coerce')

# Calculate mean and standard deviation for each group
means = df_to_analyse.groupby(['country', 'date'])[features].transform('mean')
stds = df_to_analyse.groupby(['country', 'date'])[features].transform('std')

# Avoid division by zero by replacing zeros in stds with NaN, then backfill with zeros
stds.replace(0, np.nan, inplace=True)
stds.fillna(0, inplace=True)

# Calculate Z-scores
df_to_analyse[features] = (df_to_analyse[features] - means) / stds

# Calculate the mean of the absolute Z-scores for each row
df_to_analyse['unweighted Z-score'] = df_to_analyse[features].abs().mean(axis=1)

# Display the first few rows to verify
display(df_to_analyse.head())
```

### Unweighted Z-scores

- Z-score close to 0: Indicates the audio features for that country and year are very close to the overall mean across all the data. The audio profile for that country-year is relatively average or typical.
- Positive Z-score: Indicates the audio features for that country and year have a higher mean absolute deviation from the overall mean compared to other country-years. The audio profile is more diverse or variable.
- Negative Z-score: Indicates the audio features for that country and year have a lower mean absolute deviation from the overall mean compared to other country-years. The audio profile is more homogeneous or less variable.

- (-∞; -1) : The audio profile for this country-year is very homogeneous, with low variability in the audio features compared to other country-years.
- [-1; -0.5) : The audio profile for this country-year is somewhat homogeneous, with below-average variability in the audio features.
- [-0.5; 0.5] : The audio profile for this country-year is average or typical, with variability in the audio features around the overall mean.
- (0.5; 1): The audio profile for this country-year is somewhat diverse, with above-average variability in the audio features.
- (1; +∞) : The audio profile for this country-year is very diverse, with high variability in the audio features compared to other country-years.
